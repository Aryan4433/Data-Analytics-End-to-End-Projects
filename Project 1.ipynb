```python
# rename the columns
df.columns=['State','Total_Cases','Active','Deaths','Active_Ratio','Death_Ratio']
```

```python
df.head(1)
```

```python
df.groupby(['State'])['Active_Ratio'].agg(["max"]).sort_values(by='max', ascending=False)
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State'],df['Active_Ratio'],color='purple',marker='o')

plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=13.5)

plt.xlabel('State', fontsize=15)
plt.ylabel('Active Ration', fontsize=15)

plt.title("Line Plot for Active Ratio",fontsize=15)
plt.show()
```

```python
# total cases in each state
total_cases_state= df.groupby('State')['Total_Cases'].sum().sort_values(ascending=False)
```

```python
total_cases_state
```

```python
total_cases_state.index
```

```python
plt.figure(figsize=(15, 8));
plt.xticks(rotation=75)
sns.barplot(total_cases_state.index, total_cases_state);

plt.title('Total Cases in Each State', fontsize=20);
plt.xlabel('State', fontsize=30)
plt.ylabel('Total Cases', fontsize=15);
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Top 10 state total cases
top10=total_cases_state[:10]
plt.figure(figsize=(15, 10));
plt.xticks(rotation=75)

sns.barplot(top10.index, top10);

plt.title('Total Cases in Each State', fontsize=20);
plt.xlabel('State', fontsize=30)
plt.ylabel('Total Cases', fontsize=15);
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State'],df['Active_Ratio'],color='purple',marker='o', label='Active_ratio')

plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=13.5)

plt.xlabel('State', fontsize=15)
plt.ylabel('Active Ration', fontsize=15)

plt.title("Line Plot for Active Ratio",fontsize=15)

plt.plot(df['State'],df['Death_Ratio'], color='blue',marker='*', label='deaths')
plt.legend()
plt.show()
```

```python
# pie chart shows distrinbution
# pie chart of total cases in top 5 states in india

plt.figure(figsize=(8,8))

plt.pie(x='Active', data=df.head(), labels='State', autopct='%.1f%%')
plt.title("Active Cases in Top 5 states in India", fontsize = 24) 
plt.show()

```

```python
plt.figure(figsize=(7,7))
labels = df.index
plt.pie(x='Total_Cases', data=df.head(15), labels='State', startangle=90,autopct='%.1f%%')
plt.title("Total Cases in Top 5 states in India", fontsize = 24) 
plt.tight_layout() 
plt.show()
```
