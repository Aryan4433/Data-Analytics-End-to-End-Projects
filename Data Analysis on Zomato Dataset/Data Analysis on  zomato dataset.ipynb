```python
import matplotlib as m
m.rcParams['figure.figsize']=(12,12) # is to increase figure size
sns.barplot(ratings['Aggregate rating'], ratings["Rating_counts"], data= ratings)
plt.show()
```

```python
m.rcParams['figure.figsize']=(12,12) # is to increase figure size
sns.barplot(ratings['Aggregate rating'], ratings["Rating_counts"],hue='Rating color', palette=['black','red','orange','yellow','green','#013220'] , data= ratings)
```

```python
sns.countplot(ratings['Rating color'], palette=['black','red','orange','yellow','green','#013220'] , data= ratings)
```
